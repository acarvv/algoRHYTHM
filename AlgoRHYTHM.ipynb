```python
import random

class MusicSuggestionGenerator:
    """
    Class to generate music suggestions based on a given genre.

    Attributes:
    - genre_songs: dict
        A dictionary containing different genres as keys and a list of songs as values.
    """

    def __init__(self):
        """
        Constructor to instantiate the MusicSuggestionGenerator class.

        Initializes the genre_songs attribute with some sample data.
        """

        self.genre_songs = {
            "Pop": ["Phony by Naked Cameo https://www.youtube.com/watch?v=W4UwfJ6AGPE", "Lil Boo Thang by Paul Russell https://www.youtube.com/watch?v=MoCaWpRAkVA", "Good Ones by Charli XCX https://www.youtube.com/watch?v=reIgoVVdXOg", "Till U Hate Me by Laszewo https://www.youtube.com/watch?v=Y4kzJO4H3wQ", "Bloom by Surfaces https://www.youtube.com/watch?v=TJM5Uhu84I8", "Fly by Ronis Goliath https://www.youtube.com/watch?v=MEf0AoyIcnY"],
            "Rock": ["High by Stephen Sanchez https://www.youtube.com/watch?v=DFH8FcO5Iks", "Waste A Moment by Kings of Leon https://www.youtube.com/watch?v=QAhJoXUeJ08", "Puppets by Corella https://www.youtube.com/watch?v=bkDDvdDQKec","Turn by The Night Cafe https://www.youtube.com/watch?v=L2Y5jdhI7XQ", "Chit Chat by Beach Weather https://www.youtube.com/watch?v=-flFNAegLZE"],
            "Rap": ["Nun id Change by Yeat https://www.youtube.com/watch?v=NWD5NY9nbsU" , "Like Me by $not https://www.youtube.com/watch?v=uefojCcc0r8", "Again! by Earl Stan ft. BIGBABYGUCCI https://www.youtube.com/watch?v=HlsSUPTpnYo", "Phone Numbers by Dominic Fike ft. Kenny Beats https://www.youtube.com/watch?v=ETxuFyAL1RU", "Bankroll by Nic D ft. Connor Price https://www.youtube.com/watch?v=ZggyvUXNLcE"],
            "Indie": ["Without U by Royal Otis https://www.youtube.com/watch?v=tfYiXi7a3e0", "Driving Song by The Baxbys https://www.youtube.com/watch?v=YokeUMdGvs4", "2am by the Foals https://www.youtube.com/watch?v=wfVqIjfieUU","Going Home to a Party by JW Francis https://www.youtube.com/watch?v=Zj-eXw3nYEs", "Good Girls (Don't Get Used) by Beach Bunny https://www.youtube.com/watch?v=HHrFjZH1Hfg"],
            "RnB": ["I Wished I Missed My Ex by Mahalia https://www.youtube.com/watch?v=7pKrZmLN3Ss", "Yellow of the Sun by Nao https://www.youtube.com/watch?v=eMqzyE9fIFM", "West Coast Love by Emotional Oranges https://www.youtube.com/watch?v=PaSON7HvFao", "Talk 2 U by Brent Faiyaz https://www.youtube.com/watch?v=1EVFJVYkoAY", "Nayhoo by Chon ft. Masego & lophiile https://www.youtube.com/watch?v=ztzeisJzJI8"]

        }

    def suggest_music(self, genre: str) -> str:
        """
        Generates a random music suggestion based on the given genre.

        Parameters:
        - genre: str
            The genre for which a music suggestion is to be generated.

        Returns:
        - str:
            A randomly selected song from the list of songs for the given genre.

        Raises:
        - ValueError:
            Raises an error if the given genre is not found in the genre_songs dictionary.
        """



        # Checking if the given genre exists in the genre_songs dictionary
        if genre not in self.genre_songs:
            raise ValueError("Genre not found.")

        # Selecting a random song from the list of songs for the given genre
        song = random.choice(self.genre_songs[genre])

        return song
```

**Please type one of the following genres in the box to get a song reccomendation and press enter!**

Genres: Pop, Rock, Rap, Indie or RnB (make sure the letters are capitalized exactly how it's shown here!)

```python
suggestion_generator = MusicSuggestionGenerator()

genre = input("Enter a music genre: ")
#To be able to have a bar for the person to enter in

suggestion = suggestion_generator.suggest_music(genre)
print(f"For the genre '{genre}', a suggested song is '{suggestion}'.")
```
